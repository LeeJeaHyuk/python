```python
# pr01-7 
# 특정 열을 기준으로 데이터프레임 분할 (grouping) : 특성들을 grouping한다
# 관람가를 screening_rat 컬럼 그룹핑 .groupby
df_screenRat = df.groupby('screening_rat')
# 그룹의 개수 리턴
df_screenRat.ngroups
# 키값 확인
df_screenRat.groups.keys()
```

```python
# pr01-7 
# 그룹핑한 특정 키값의 values 리턴
df_screenRat.get_group('12세 관람가').head(2)
```

```python
# pr01-7 
# 그룹핑한 후 집계함수 사용
# 각 그룹의 개수를 구해보자
df_screenRat['genre'].count()
# 각 그룹의 관객수를 구해보자
df_screenRat['box_off_num'].sum()
# 각 그룹의 최대 관객수를 구해보자
df_screenRat['box_off_num'].max()
# df_screenRat['box_off_num'].min()
# 최대 상영 시간
# df_screenRat['time'].max()
```

```python
# pr01-7 
# GroupBy.agg(func, *args, **kwargs) 이용해서 연산
# 각 그룹에서 'time','box_off_num'의 최대값을 구해보자
df_screenRat.agg('max')[['time','box_off_num']]
df_screenRat.agg({'time': min, 'box_off_num': max}) #각각 다르게 지정할 수 있다 -> indexing을 해 주어야 할 것 같다
```

```python
# pr01-7 
# 각 그룹 별로 	time	box_off_num의 최대값과 최솟값을 출력
df_screenRat.agg({'time': [max,min], 'box_off_num': [max,min]})
```

```python
# pr01-7 
# GroupBy.agg(func, *args, **kwargs)이용해서 연산
# 사용자 정의 함수 time 컬럼의 데이터들 중에 가장 시간이 긴/짧은 시간의 차를 추출

def my_Func(x):
    diff = x.max()-x.min()
    return(diff)

df_screenRat[['time','box_off_num']].agg(my_Func)
```

```python
# pr01-8 
# 시계열 데이터 사용
# df 컬럼을 시계열 객체로 변환
# 'release_time' 열을 datetime 형식으로 바꿈
df['my_date']=pd.to_datetime(df['release_time'])
# my_date의 type을 datetime64[ns]임을 볼 수 있다
df.info()
```

```python
# pr01-8 
# my_date 의 컬럼을 이용해서 년, 월, 일로 분철해서 각각의 컬럼을 추가하자
# df[].dt.year
df['year'] = df['my_date'].dt.year
df['month'] = df['my_date'].dt.month
df['day'] = df['my_date'].dt.day
df['dayofweek'] = df['my_date'].dt.dayofweek
df.info()
```

```python
# pr01-8 
# 년도별 최대 관객수
df.groupby('year')['box_off_num'].max()
# 월별 최대 관객수
df.groupby('month')['box_off_num'].max()
```

```python
# pr01-9
# 누락데이터 확인
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
titanic['deck'].value_counts()
titanic['deck'].value_counts(dropna = False)
```

```python
# pr01-9
# 누락데이터 확인 2
titanic.isnull().sum()
```

```python
# pr01-9
# 누락데이터 제거 객체
# DataFrame.dropna(axis=0, how=_NoDefault.no_default, thresh=_NoDefault.no_default, subset=None, inplace=False)
# axis = 0 행 전체 삭제 axis = 1 열 전체 삭제
print(f'row의 개수 : {len(titanic)}')

# row(891)-thresh(400) = null값이 491보다 많으면 삭제
res = titanic.dropna(axis=1,thresh = 400)
titanic.isnull().sum()
res.isnull().sum()

# row(891)-thresh(800) = null값이 91보다 많으면 삭제
res00 = titanic.dropna(axis=1,thresh = 800)
res00.isnull().sum()
```

```python
# pr01 -9
# axis=0 행 전체 삭제 thresh=n
# thresh=n 보다 null이 아닌값이 적으면 행 삭제
print(f'column의 개수 : {len(titanic.columns)}')

# 총 컬럼 15 thresh값 14로 설정했으므로 행에서 null값이 2개이상이면 행이 삭제된다
# column(15)-thresh(14) = 1
res02 = titanic.dropna(axis=0,thresh = 14)
titanic.isnull().sum()
res02.isnull().sum()
```

```python
# pr01 -9
# 누락 데이터 치환
age_mean = titanic['age'].mean()
age_mean

# .fillna(치환할 값, inplace = )
# titanic['age'].fillna(age_mean, inplace = True)
titanic['age'].fillna(titanic['age'].mean(), inplace = True)
titanic.isnull().sum()
```

```python
# pr01 -9
# embarked embark_town의 최빈값 구해서 그 값으로 채우기
titanic.groupby('embarked')['embarked'].count()
titanic.groupby('embark_town')['embark_town'].count()

# Southampton 이 최빈값으로 보인다.
# idxmax()  idxmin() / 최빈값 최소값


e_idxmax = titanic['embarked'].value_counts(dropna=False).idxmax()
e_idxmax
titanic['embarked'].fillna(e_idxmax, inplace=True)

et_idxmax = titanic['embark_town'].value_counts(dropna=False).idxmax()
titanic['embark_town'].fillna(et_idxmax, inplace=True)

titanic.isnull().sum()
```
